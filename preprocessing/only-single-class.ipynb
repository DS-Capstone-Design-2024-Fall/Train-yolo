```python
import os 

root_path = "D:/Downloads/street-facilities-2/labels"
save_root_path = "D:/Downloads/street-facilities-2/labels-only0"


sub_path = ["train", "val", "test"]

cnt_classes = [0, 0, 0]

for sub in sub_path:
    label_path = os.path.join(root_path, sub)
    save_label_path = os.path.join(save_root_path, sub)
    os.makedirs(save_label_path, exist_ok=True) # 없는 폴더면 새로 생성
    print(label_path)

    labels = [label for label in os.listdir(label_path) if label.endswith('.txt')]
    total_cnt = len(labels)

    for i, label in enumerate(labels):
        if i % 300 == 0 :
            print(f"{i+1} / {total_cnt}")
        annotations = []
        with open(os.path.join(label_path, label), "r+", encoding="utf-8") as f:
            # 기존 내용 읽기 및 수정
            for line in f:
                elems = line.split(" ")
                # 변경 전 클래스 분포 집계
                cnt_classes[int(float(elems[0]))] += 1
                # class 0(braille-block-defect), 1(sidewalk-block-defect), 2(bicycle-road-defect) 
                # -> 모두 class 0 (crack) 으로 통일
                elems[0] = str(0) 
                newline = " ".join(elems)
                annotations.append(newline)
                
            # # 새 내용 작성 (덮어쓰기)
            # f.seek(0) # 파일 포인터 이동
            # for annotation in annotations:
            #     f.write(annotation)
            # f.truncate() # 파일 포인터 이전 내용만 남기고 나머지 삭제

        # 새 내용 작성 (별도 파일에 작성)
        with open(os.path.join(save_label_path, label), "w", encoding='utf-8') as f:
            f.seek(0) # 파일 포인터 이동
            for annotation in annotations:
                f.write(annotation)


print("변경 전 클래스 분포")
print(cnt_classes)
```
